```python
table = pd.DataFrame(results)
table
```

```python
# Create TF-IDF Vocab
```

```python
tfidf_counter = TfidfVectorizer(analyzer="word", stop_words=stopwords.words("english"))
docs = []
for i in results:
    for key in i.keys():
        if key == "text":
            docs.append(i[key])

tfidf_results = tfidf_counter.fit_transform(docs).toarray()
vocab = tfidf_counter.get_feature_names()

keyWords = {};
for j in range(len(tfidf_results)):
    subject = results[j]['subject']
    if 'subject' not in keyWords.keys():
        keyWords[subject] = [];
    for i in range(len(tfidf_results[j])):
        s = tfidf_results[j][i]
        if s > threshold:
            if vocab[i] not in keyWords[subject]:
                keyWords[subject].append(vocab[i])

```

```python
ideal_text_books = {};
for j in range(len(results)):
    subject = results[j]['subject']
    if 'subject' not in ideal_text_books.keys():
        ideal_text_books[subject] = '';  
    for sentence in results[j]['sentences']:
        for word in keyWords[subject]:
            added = false
            if word in sentence and not added:
                ideal_text_books[subject] += ' ' + sentence
                added = true
```
